```python
#emisssion matrix
def count_pairs(word_list,label):
  result={}
  for i,j in zip(word_list,label):
    pair=(i,j)
    if pair in result:
      result[pair]+=1
    else:
      result[pair] = 1
  return result
    
```

```python
result=count_pairs(word_list,label)
```

```python
result
```

```python
# emission probability
def emission_proability(result,total_values):
  temp_dict={}
  for i in result:
    temp_dict[i]=result[i]/total_values[i[1]]
  return temp_dict


```
