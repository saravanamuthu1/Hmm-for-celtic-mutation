```python
#emisssion matrix
def count_pairs(word_list,label):
  result={}
  for i,j in zip(word_list,label):
    pair=(i,j)
    if pair in result:
      result[pair]+=1
    else:
      result[pair] = 1
  return result
    
```

```python
# emission probability
def emission_proability(result,total_values):
  temp_dict={}
  for i in result:
    temp_dict[i]=result[i]/total_values[i[1]]
  return temp_dict


```

```python
def trans_value(tag2,tag1,train_bag):
  tags=[]
  for i in train_bag:
    tags.append(i[1])
  count1= (len(list(filter(lambda x: x == tag1, tags))))
  count2=0
  for index in range(0,len(tags)-1):
    if tags[index] == tag1 and  tags[index+1] == tag2:
      count2+=1
  return (count2,count1)
  
```

```python
trans_matrix=np.zeros((len(freq1),len(freq1)),dtype='float32')


```

```python
for i,tag1 in enumerate(freq1):
  for j,tag2 in enumerate(freq1):
    try:
      trans_matrix[i,j]=trans_value(tag2,tag1,train_bag)[0]/trans_value(tag2,tag1,train_bag)[1]
    except ZeroDivisionError:
      trans_matrix[i,j] = 0.0
```

```python
trans_matrix
```

```python

tags_df = pd.DataFrame(trans_matrix, columns = freq1, index=freq1)
```

```python
tags_df
```

```python
def Vertibi(words,train_bag):
  state=[]
  tags=list(set([pair[1] for pair in train_bag]))
  for key,word in enumerate(words):
    p=[]
    for i in tags:
      pair=(word,i)
      if key == 0:
        transmission_prob=tags_df.loc[i,i]
      else:
        transmission_prob=tags_df.loc[state[-1],i]
      if pair in emission_matrix.keys():
        emission_p=emission_matrix[pair]
      else:
        emission_p=0.0
      state_prop=emission_p*transmission_prob
      p.append(state_prop)
    max_value=max(p)
    value=tags[p.index(max_value)]
    state.append(value)
    print(len(state))
  return list(zip(words,state))

```

```python
value=Vertibi(test_word,train_bag)
```
